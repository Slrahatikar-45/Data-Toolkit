Data Toolkit

1.What is NumPy, and why is it widely used in Python?
- NumPy (Numerical Python) is a fundamental package for numerical computing in Python. It provides support for large multidimensional arrays and matrices, along with a collection of mathematical functions to operate on them efficiently.
- It's widely used for data science, machine learning, and scientific computing.

2.How does broadcasting work in NumPy?
- Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. It stretches the smaller array across the larger one without copying data, enabling vectorized operations and improving performance.

3.What is a Pandas DataFrame?
- A DataFrame is a 2D labeled data structure with columns of potentially different data types. It’s like a spreadsheet or SQL table in Python and is ideal for data manipulation and analysis.

4.Explain the use of the groupby() method in Pandas.
- The groupby() method splits data into groups based on a specified key, applies a function to each group, and combines the results. It's commonly used for aggregation, transformation, and filtering tasks.

5.Why is Seaborn preferred for statistical visualizations?
- Seaborn is built on top of Matplotlib and simplifies complex visualizations with fewer lines of code. It provides beautiful default styles and built-in themes for better aesthetic and statistical plots.

6.What are the differences between NumPy arrays and Python lists?
- NumPy arrays are more efficient in memory and speed, support vectorized operations, and are limited to a single data type. Python lists are more flexible but slower and support mixed data types.

7.What is a heatmap, and when should it be used?
- A heatmap is a data visualization that shows the magnitude of a phenomenon using color. It is useful for visualizing correlation matrices, missing values, or any matrix-like data.

8.What does the term “vectorized operation” mean in NumPy?
- Vectorized operations perform computations on entire arrays without explicit loops, making the code faster and more concise. NumPy achieves this using underlying C-based operations.

9.How does Matplotlib differ from Plotly?
- Matplotlib is static and primarily used for 2D plots, while Plotly provides interactive, web-based charts. Plotly is better for dashboards and user engagement, while Matplotlib is often used for quick prototyping.

10.What is the significance of hierarchical indexing in Pandas?
- Hierarchical indexing (MultiIndex) allows multiple levels of indexing in a DataFrame or Series. It enables complex data structures and easier data manipulation for multi-dimensional datasets.

11.What is the role of Seaborn’s pairplot() function?
- pairplot() visualizes pairwise relationships between numeric variables in a dataset. It's useful for exploring patterns, trends, and correlations among multiple features.

12.What is the purpose of the describe() function in Pandas?
- The describe() function provides summary statistics like mean, std, min, max, and percentiles for numerical columns, helping with quick exploratory data analysis.

13.Why is handling missing data important in Pandas?
- Missing data can skew analysis, lead to errors, or reduce model performance. Pandas provides tools to detect, remove, or impute missing values to maintain data integrity.

14.What are the benefits of using Plotly for data visualization?
- Plotly supports interactive plots, is easy to integrate with web apps, and provides 3D plotting and real-time data visualizations. It’s highly customizable and works well with large datasets.

15.How does NumPy handle multidimensional arrays?
- NumPy supports n-dimensional arrays (ndarrays) and allows efficient operations like reshaping, slicing, and broadcasting. It also supports matrix operations, transposition, and tensor manipulations.

16.What is the role of Bokeh in data visualization?
- Bokeh creates interactive and real-time visualizations for the web. It's useful for dashboards, streaming data, and is often used with Flask or Django for embedding visualizations.

17.Explain the difference between apply() and map() in Pandas.
- map() is used with Series and applies a function element-wise. apply() can be used with both Series and DataFrames to apply a function along rows or columns, offering more flexibility.

18.What are some advanced features of NumPy?
- Advanced features include broadcasting, universal functions (ufuncs), linear algebra operations, Fourier transforms, random number generation, and masked arrays.

19.How does Pandas simplify time series analysis?
- Pandas supports datetime indexing, resampling, frequency conversion, and moving window statistics, making time-based data manipulation and analysis intuitive and powerful.

20.What is the role of a pivot table in Pandas?
- A pivot table summarizes data by grouping and aggregating values across two axes. It is useful for quickly analyzing and comparing subsets of data.

21.Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy uses contiguous memory blocks and optimized C code, making array slicing faster and more efficient. Python lists store references, which makes slicing slower.

22.What are some common use cases for Seaborn?
- Seaborn is commonly used for visualizing distributions, relationships between variables, heatmaps, regression analysis, and statistical comparisons with minimal code.





```python
#1. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print("Sum of each row:", row_sum)
```

```python
#2.Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [5, 15, 25]})
mean_a = df['A'].mean()
print("Mean of column A:", mean_a)
```

```python
#3.Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns

df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [2, 4, 6, 8]})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#5.Generate a bar plot using Plotly.

import plotly.express as px

data = {'Fruits': ['Apple', 'Banana', 'Mango'], 'Count': [10, 15, 7]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Fruits', y='Count', title="Fruit Count")
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.

df = pd.DataFrame({'A': [1, 2, 3], 'B': [10, 20, 30]})
df['C'] = df['A'] + df['B']
print(df)
```

```python
#7.Write a program to perform element-wise multiplication of two NumPy arrays.

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = np.multiply(arr1, arr2)
print("Element-wise multiplication:", result)
```

```python
#8.Create a line plot with multiple lines using Matplotlib.

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 7]
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = some function')
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
```

```python
#9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

df = pd.DataFrame({'Scores': [45, 67, 89, 30]})
filtered_df = df[df['Scores'] > 50]
print(filtered_df)
```

```python
#10.Create a histogram using Seaborn to visualize a distribution.

data = [10, 20, 20, 30, 30, 30, 40]
sns.histplot(data, bins=5, kde=True)
plt.title("Histogram")
plt.show()
```

```python
#11.Perform matrix multiplication using NumPy.

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.matmul(a, b)
print("Matrix multiplication result:\n", result)
```

```python
#13.Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(
    x=[1, 2, 3], y=[4, 5, 6], z=[7, 8, 9],
    mode='markers',
    marker=dict(size=8, color='blue')
)])
fig.update_layout(title="3D Scatter Plot")
fig.show()
```
